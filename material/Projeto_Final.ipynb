```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('mapbiomas.csv', delimiter=';', encoding='latin1')
```

```python
df.columns = df.columns.str.replace('ï»¿', '', regex=False)
```

```python
df.columns = df.columns.str.strip()
```

```python
df.columns = df.columns.str.replace('Ã§', 'ç').str.replace('Ã£', 'ã').str.replace('Ã¡', 'á').str.replace('Ã', 'Á').str.replace('Ã³' , 'ó')
```

```python
df.head(10000)
```

```python
df['Estados'] = df['Estados'].replace('RondÃ´nia', 'Rondônia')
```

```python
df.head(10000)
```

```python
for column in df.columns:
    if df[column].dtype == 'object' and column not in ['Ano', 'Estados']:
        df[column] = df[column].str.replace(',', '.').str.replace(' ', '').astype(float)
```

```python
dados_1985 = df[df['Ano'] == 1985]
soma_area_vegetada_1985 = dados_1985['Floresta'].sum()
soma_area_agropecuaria_1985 = dados_1985['Agropecuária'].sum()
```

```python
print(f"Soma da Área Vegetada em 1985: {soma_area_vegetada_1985:.2f} ha")
print(f"Soma da Área Agropecuária em 1985: {soma_area_agropecuaria_1985:.2f} ha")
```

```python
dados_2023 = df[df['Ano'] == 2023]
soma_area_vegetada = dados_2023['Floresta'].sum()
soma_area_agropecuaria = dados_2023['Agropecuária'].sum()
print(f"Soma da Área Vegetada em 2023: {soma_area_vegetada:.2f} ha")
print(f"Soma da Área Agropecuária em 2023: {soma_area_agropecuaria:.2f} ha")
```

```python
soma_area_floresta_km2 = soma_area_vegetada * 0.01
soma_area_agropecuaria_km2 = soma_area_agropecuaria * 0.01
print(f"Soma da Área Vegetada em km²: {soma_area_floresta_km2:.2f} km²")
print(f"Soma da Área Agropecuária em km²: {soma_area_agropecuaria_km2:.2f} km²")
```

```python
soma_area_floresta_km2_1985 = soma_area_vegetada_1985 * 0.01
soma_area_agropecuaria_km2_1985 = soma_area_agropecuaria_1985 * 0.01
print(f"Soma da Área Vegetada em km² em 1985: {soma_area_floresta_km2_1985:.2f} km²")
print(f"Soma da Área Agropecuária em km² em 1985: {soma_area_agropecuaria_km2_1985:.2f} km²")
```

```python
anos = ['1985', '2023']
areas_floresta = [soma_area_floresta_km2_1985, soma_area_floresta_km2]
areas_agropecuaria = [soma_area_agropecuaria_km2_1985, soma_area_agropecuaria_km2]
bar_width = 0.20
x = range(len(anos))
fig, ax = plt.subplots(figsize=(10, 5))
bar1 = ax.bar(x, areas_floresta, width=bar_width, label='Área Vegetada (Floresta)', color='#70a145', alpha=1)
bar2 = ax.bar([p + bar_width for p in x], areas_agropecuaria, width=bar_width, label='Área Agropecuária', color='#d8a155', alpha=1)
ax.set_xlabel('Ano')
ax.set_ylabel('Área (km²)')
ax.set_title('Comparação da Área Vegetada e Agropecuária em 1985 e 2023 no Acre')
ax.set_xticks([p + bar_width / 2 for p in x])
ax.set_xticklabels(anos)
ax.legend(title='Uso do Solo')
ax.grid(axis='y')
plt.tight_layout()
plt.show()
```
